```python
dispopivot = pd.pivot(df_dispoalim, index=['Zone'], columns=['Élément'],values=['Valeur'])
dispopivot
```

```python
dispopivot.isna().sum()
```

Je vais remplacer les valeurs manquantes de la colonne 'Exportations' en utilisant cette formule:  

Disponibilité Intérieure = (Production + Importations - disponibilité alimentaire(Quantité) + Variation de stock) - Exportations
devient alors:  
**Exportations = (Production + Importations - disponibilité alimentaire(Quantité) + Variation de stock) - Disponibilité Intérieure**

```python
dispopivot['Exportations - Quantité']= dispopivot.apply(lambda x : ((x['Production'] + x['Importations - Quantité'] 
                                                                       - x['Disponibilité alimentaire en quantité (kg/personne/an)'] 
                                                                       + x['Variation de stock']) - x['Disponibilité intérieure']) 
                                                          if pd.isnull(x['Exportations - Quantité']) else x['Exportations - Quantité'], axis=1)
```

```python
dispopivot.isna().sum()
```

```python
#nouvelles variables
df_final['Tx dépendance importation'] = df_final['Importations - Quantité'] / (df_final['Production'] + df_final['Importations - Quantité'] - df_final['Exportations - Quantité']) * 100
df_final['Tx d autosuffisance'] = df_final['Production'] / (df_final['Production'] + df_final['Importations - Quantité'] - df_final['Exportations - Quantité']) * 100
df_final
```

```python
df_final=df_final.drop(['Disponibilité alimentaire (Kcal/personne/jour)','Disponibilité alimentaire en quantité (kg/personne/an)', 'Nourriture', 'Production', 'Variation de stock'], axis=1)
df_final
```

```python
#définition de l'index pour utiliser la méthode PCA pour identifier les outliers
df_final.set_index('Zone', inplace = True) 
```

```python
# Sélectionner les colonnes numériques
cols_num = df_final.select_dtypes(include=['float64', 'int64']).columns
df_num = df_final[cols_num]

# Standardiser les données
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_num)

# Appliquer la méthode PCA
pca = PCA(n_components=9)
pca.fit(df_scaled)
principal_components = pca.transform(df_scaled)

# Identifier les outliers
outliers = df_final.iloc[principal_components[:, 0].argsort()[-3:][::-1]]
outliers
```

```python
#suppression des 3 outliers pour ne pas biaiser notre analyse
# Supprimer les outliers du DataFrame final
df_final = df_final.drop(outliers.index)
df_final.describe()
```
